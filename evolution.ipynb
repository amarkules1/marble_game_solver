```python
from main.game import Game
from main.move_selectors import run_generation, games_to_df, survivor_ct_function_1, survivor_ct_function_2, children_per_survivor_func_1, survivor_ct_function_3, children_per_survivor_func_2
import pandas as pd
```

```python
init_game = Game()
```

```python
games = [init_game]
```

```python
finished_games = pd.DataFrame(columns = ['score', 'remaining_marbles', 'moves'])
```

```python
for i in range(0,100):
    games = [init_game]
    while len(games) > 0:
        games = run_generation(games, survivor_ct_function_1, children_per_survivor_func_1)
        complete_games = list(filter(lambda game: game.is_finished(), games))
        games = list(set(games) - set(finished_games))
        finished_games = pd.concat([finished_games, games_to_df(complete_games)], ignore_index=True)

finished_games.to_csv("evolution_v1_results.csv")
```
